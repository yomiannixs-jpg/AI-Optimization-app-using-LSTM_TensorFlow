```python
!pip install -q yfinance

import yfinance as yf
import pandas as pd
from datetime import datetime

# Define tickers
tickers = ['CL=F', 'BZ=F', '^GSPC', 'URTH', '^VIX', '^OVX', 'GC=F', 'TLT', 'LQD', 'DX-Y.NYB']
names = ['WTI', 'Brent', 'S&P500', 'MSCI', 'VIX', 'OVX', 'Gold', 'Treasury', 'CorpBond', 'Dollar']

# Download data
df = yf.download(tickers, start='2020-01-01', end='2026-06-27', auto_adjust=True)

# Save close prices
close_prices = df['Close']
close_prices.columns = names
close_prices.to_csv('all_assets.csv')

# Download to computer
from google.colab import files
files.download('all_assets.csv')

print("Download complete!")
print(close_prices.head())
```
